```python
cost_level = df[["Member_number", "itemDescription"]].sort_values(by = "Member_number", ascending = False)
cost_level["itemDescription"] = cost_level["itemDescription"].str.strip()
cost_level
```

```python
transactions = [a[1]['itemDescription'].tolist() for a in list(cost_level.groupby(["Member_number"]))]
```

```python
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.002,min_confidence = 0.05, min_lift = 3, min_length = 2, max_length=4)
```

```python
results = list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs = [tuple(result[2][0][0])[0] for result in results]
    rhs = [tuple(result[2][0][1])[0] for result in results]
    supports = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsindataframe = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidences', 'Lifts'])
```

```python
resultsindataframe.nlargest(n=20, columns = "Lifts")
```
